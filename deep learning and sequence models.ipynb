classification of neural network 
-shallow neural network(has on hidden layer between the input and output layer)

-deep neural network(has more than one hidden layer between the input and 
output layer)

Types of deep neural newtork are:
1.Feed-forward neural network
   -its the simplest and has the flow in one direction,meaning the flow stars at the 
   input then to the hidden layer the to the outputlayer,this means there is no
   loop in this neural network thus the flow of info stops at the output layer.

2.Recurrent neural network
   -its a multilayer that its data flow in sequences thus it involves loops.
Common uses of RNN
-Help securities traders to generate analytic reports
-Detect abnormalities in the contract of financial statement
-Detect fraudulent credit-card transaction
-Provide a caption for images
-Power chatbots
-The standard uses of RNN occur when the practitioners are working with 
time-series data or sequences (e.g., audio recordings or text).

3.Convolution neural network 
   -its used for more complex features mostly used for unstructed dataset 
and the practitioner need to extract information from it.
consider the following example:
For instance, if the task is to predict an image caption:
1.The CNN receives an image of let's say a cat, this image, in computer term, 
 is a collection of the pixel. Generally, one layer for the greyscale picture 
 and three layers for a color picture.
2.During the feature learning (i.e., hidden layers), the network will identify 
  unique features, for instance, the tail of the cat, the ear, etc.
3.When the network thoroughly learned how to recognize a picture, it can provide
  a probability for each image it knows. The label with the highest probability 
  will become the prediction of the network.
